```python
import numpy as np
a = np.arange(15).reshape(3, 5)
print(a)
print('type(a)=',type(a))  # type(a)= <class 'numpy.ndarray'>
print('a.ndim=',a.ndim)    # 陣列有多少維度
print('a.shape=',a.shape)  # 每個維度的大小
print('a.size=',a.size)    # 陣列當中有幾個元素
print('a.dtype=',a.dtype)  #  陣列中的資料型態
print('a.itemsize=',a.itemsize)  # 陣列中每個元素佔用的空間
print('a.data=',a.data)    # 陣列所存在的記憶體位置
```

### 陣列的元素會配置在連續的記憶體位置，而列表則沒有這樣的限制。


```python
# List 與 Array 的轉換 : tolist()
# ndarray to list --------------------
a = np.arange(15).reshape(3, 5)
print('a=\n',a)
print('list(a) = \n',list(a))
print("-----------------------------------------------------")
print('a.tolist()=\n',a.tolist())
tolist_test = []
for i in  list(a):
    trmplist =list(i)
    tolist_test.append(trmplist)
print(tolist_test)
```

### 應該使用 Numpy 的原因 :
* Numpy 使用更少的空間儲存資料
* Numpy 對於元素的查找更加快速高效
* Numpy 可以更好地表示數學中的向量/矩陣
* Numpy 支援了矩陣相關的數學運算

```python
# numpy 中的型態判斷
print(a.dtype == 'int64') 
print(a.dtype is 'int64') 
```

```python
# 資料型態的正確用法
print(a.dtype == 'int64') # True
print(a.dtype is 'int64') # False
print(a.dtype is np.dtype('int64')) # True
print(a.dtype == 'int') 
print(a.dtype == np.int) 
print(a.dtype == np.dtype('int')) 
```
